```python
import pandas as pd
import numpy as np

# Creating a sample dataset
data = {
    'Year': [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020],
    'Income': [50000, 54000, 55000, 56000, 57000, 60000, 63000, 66000, 68000, 68000, 70000]
}

# Creating DataFrame
df = pd.DataFrame(data)

# Display the DataFrame
print(df)
```

```python
import matplotlib.pyplot as plt

from sklearn import linear_model
```

```python
plt.scatter(df['Year'], df['Income'], color = 'blue' , marker = '+')
plt.xlabel('Year')
plt.ylabel('Income')
plt.show()
```

```python
linear_model = linear_model.LinearRegression()
reg = linear_model.fit(df[['Year']], df['Income'])
```

```python
prediction = reg.predict(df[['Year']])
```

```python
plt.scatter(df['Year'], df['Income'], color='red', marker='+')
plt.xlabel('Year')
plt.ylabel('Income')
plt.plot(df['Year'], prediction, color = 'blue')
plt.show()
```

```python
reg.predict([[2021]])
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(df['Income'], prediction)
print(mse)
rmse = np.sqrt(mse)
print(rmse)
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(df['Income'], prediction)
print(r2)
```
