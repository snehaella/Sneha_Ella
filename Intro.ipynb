```python
# Matplotlib Basics

import numpy as np
import matplotlib.pyplot as plt
```

```python
# line plots

x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot Example')
plt.show()
```

```python
# scatter plots: useful for showing the relationship between two variables


x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.scatter(x, y, color='green')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot Example')
plt.show()
```

```python
# Bar charts : ideal for comparing different categories

categories = ['A', 'B', 'C', 'D']
values = [4, 7, 1, 8]

plt.bar(categories, values, color='yellow')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart Example')
plt.show()
```

```python
# intro to seaborn
#Box plot: used to show the distribution of data and identify outliers

import seaborn as sns
import matplotlib.pyplot as plt

#data = sns.load_dataset('tips')
#sns.scatterplot(x='total_bill', y='tip', data=data)
data = sns.load_dataset('iris')
sns.boxplot(x='species', y='petal_length', data=data, palette='Set2') #palette change the color its not neccesary
plt.xlabel('Species')
plt.ylabel('Petal Length')
plt.title('Box Plot Example')


plt.show()
```

```python
# Violin plots: combine the feature of box plots and density plots

data = sns.load_dataset('iris')
sns.violinplot(x='species', y='petal_length', data=data, palette='Set3') #palette change the color its not neccesary
plt.title('Violin Plot Example')
plt.show()
```

```python
# Heatmaps : visualize matrix like data

import seaborn as sns
import matplotlib.pyplot as plt

data = sns.load_dataset('flights')
data_pivot = data.pivot(index='month', columns='year', values='passengers')
sns.heatmap(data_pivot, annot=True, fmt='d', cmap='coolwarm') #cmap change the color its not neccesary
plt.title('Heatmap Example')
plt.show()
```
